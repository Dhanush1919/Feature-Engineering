#### Univariate selection:

```python
X = df.iloc[:,:-1] #Not considering the last column alone
Y = df['price_range']
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
ordered_rank_features = SelectKBest(score_func=chi2,k=20)
ordered_features = ordered_rank_features.fit(X,Y)
```

```python
ordered_features
```

```python
ordered_features.scores_
```

```python
Scores = pd.DataFrame(ordered_features.scores_)
Feature = pd.DataFrame(X.columns)
```

```python
Rank_values = pd.concat([Feature,Scores],axis=1)
```

```python
Rank_values.columns = ['Feature_name','Score']
```

```python
Rank_values
```

```python
# Top 10 important features:
Rank_values.nlargest(10,'Score')
```

### Feature Importance :
This technique gives a score to each feature , higher the score , more relevant it is

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model = ExtraTreesClassifier()
model.fit(X,Y)
print(model.feature_importances_)
```

```python
# Visualizing the series:
Feature_series = pd.Series(model.feature_importances_,X.columns)
Feature_series.nlargest(10).plot(kind='barh')
plt.show()
```

```python
dataset = df.iloc[:,:-1].corr()
len(dataset.columns)
```

```python
# Getting the highly correlated columns: (checks lower triangular) 
def correlation(dataset,threshold):
    corr_matrix = dataset.corr()
    col_name = set()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                cols = corr_matrix.columns[i]
                col_name.add(cols)
    return col_name
```

```python
correlation(df.iloc[:,:-1],0.5)
```

### Information gain:

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
values = mutual_info_classif(X,Y)
```

```python
top_features = pd.Series(values,index=X.columns)
```

```python
# The Features with higher value is imp , while the feature with lower value or 0 can be ignored:
top_features.sort_values(ascending=False)
```
