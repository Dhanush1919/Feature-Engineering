```python
## pip install pystan
## conda install -c conda-forge fbprophet
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pystan
import fbprophet
```

#### Data preprocessing:

```python
df.drop(168,axis=0,inplace=True)
```

```python
df.plot()
```

```python
#Renaming the columns:
df.columns = ['ds','y'] 
```

```python
df.head()
```

```python
#Converting into datetime format:
df['ds'] = pd.to_datetime(df['ds'])
```

```python
df.head()
```

```python
df['y'].shift(1)
```

```python
# Replacing the difference with the same value: Shifting to make the data stationary
# Done to make the data have constant variance and mean throughout the series
#df['y'] = df['y']-df['y'].shift(1)
```

```python
df['y'].head()
```

```python
df['y'].plot()
```

```python
from fbprophet import Prophet
```

```python
# Inbuilt functions available in prophet:
dir(Prophet)
```

#### Initalizing the model:

```python
p = Prophet()
#p.fit(df)
```

```python
p.fit(df)
```

```python
p.seasonalities
```

```python
### Cfreating future dates , appending the present df with next 365 days:
future_date = p.make_future_dataframe(periods=365)
```

```python
future_date
```

```python
# Making predictions:
Prediction = p.predict(future_date)
Prediction.head()
```

```python
#yhat_lower and upper represents the range within yhat occurs:
Prediction[['ds','yhat','yhat_lower','yhat_upper']].head(10)
```

The blue line in the Graph represents the Predictions and the Black points represents the data in
the Dataset 

```python
# Plotting the predictions:
p.plot(Prediction)
```

```python
#Visualize each components:
p.plot_components(Prediction)
```

### Cross validation

```python
from fbprophet.diagnostics import cross_validation
```

```python
df_cv = cross_validation(p,horizon="365 days",period="180 days",initial='1095 days')
```

```python
df_cv.head()
```

```python
df_cv.shape
```

```python
from fbprophet.diagnostics import performance_metrics
```

```python
df_performance = performance_metrics(df_cv)
```

```python
#Horizon:
df_performance.head()
```

```python
from fbprophet.plot import plot_cross_validation_metric 
```

```python
fig = plot_cross_validation_metric(df_cv,metric='rmse')
```
